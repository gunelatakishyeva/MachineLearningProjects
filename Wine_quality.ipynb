```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/WineQT.csv')
df
```

```python
df=df.drop('Id',axis=1)
```

```python
df
```

```python
df.info()
```

```python
X=df.drop('quality',axis=1)
y=df['quality'].copy()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,stratify=y,test_size=0.2,random_state=42)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
num_features=X_train.select_dtypes(include=[np.number]).columns
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline=Pipeline([
    ('scaler',StandardScaler())
  ])

transformer=ColumnTransformer([
    ('num_pipeline',num_pipeline,num_features)
])
```

```python
X_transformed_train=transformer.fit_transform(X_train)
transformed_columns = transformer.get_feature_names_out(num_features)
transformed_columns
```

```python
X_transformed_train_df=pd.DataFrame(X_transformed_train,columns=transformed_columns)
X_transformed_train_df
```

```python
from sklearn.svm import SVC
```

```python
svm=SVC()
```

```python
svm.fit(X_transformed_train,y_train)
```

```python
X_transformed_test=transformer.transform(X_test)
transformed_columns = transformer.named_transformers_['num_pipeline'].named_steps['scaler'].get_feature_names_out(num_features)
transformed_columns
```

```python
svm.score(X_transformed_test,y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

rnd=RandomForestClassifier()
rnd.fit(X_transformed_train,y_train)
```

```python
rnd.score(X_transformed_test,y_test)
```

```python
from sklearn.metrics import f1_score
```

```python
y_pred=svm.predict(X_transformed_test)
```

```python
f1_score(y_test,y_pred,average='micro')
```
