```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/insurance.csv')
df
```

```python
df['age'].value_counts()
```

```python
import matplotlib.pyplot as plt
```

```python
X=df.drop('charges',axis=1)
y=df['charges'].copy()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train
```

```python
num_features=X_train.select_dtypes(include=[np.number]).columns
cat_features= X_train.select_dtypes(exclude=[np.number]).columns
```

```python
num_features
```

```python
cat_features
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='median'))
])
cat_pipeline=Pipeline([
    ('impute',SimpleImputer(strategy='most_frequent')),
    ('onehot',OneHotEncoder())
])

transformer=ColumnTransformer([
    ('num_pipeline',num_pipeline,num_features),
    ('cat_pipeline',cat_pipeline,cat_features)
])
```

```python
X_transformed_train=transformer.fit_transform(X_train)
transformed_columns=transformer.get_feature_names_out()
transformed_columns
```

```python
X_transformed_train_df=pd.DataFrame(X_transformed_train,columns=transformed_columns)
X_transformed_train_df
```

```python
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression()
lin_reg.fit(X_transformed_train,y_train)
```

```python
X_transformed_test=transformer.transform(X_test)
transformed_columns=transformer.get_feature_names_out()
transformed_columns
```

```python
lin_reg.score(X_transformed_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures

poly_features=PolynomialFeatures(degree=2, include_bias=False)
X_poly=poly_features.fit_transform(X_transformed_train)
```

```python
lin_reg=LinearRegression()
lin_reg.fit(X_poly,y_train)
lin_reg.intercept_,lin_reg.coef_
```

```python
X_poly_test=poly_features.transform(X_transformed_test)
```

```python
lin_reg.score(X_poly_test,y_test)
```
